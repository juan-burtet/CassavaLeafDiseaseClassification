```python
from fastai.vision.all import *
```

```python
root_path = '/kaggle/input/cassava-leaf-disease-classification/'
df_train = pd.read_csv(root_path + 'train.csv')
dls = ImageDataLoaders.from_df(df_train, path=root_path+"train_images/", item_tfms=Resize(224))
```

```python
path, name, model = ('models-analise', 'resnet152_tune_10', resnet152)

learn = cnn_learner(dls, model, metrics=[error_rate,accuracy], pretrained=False)
```

```python
learn.model_dir = '/kaggle/input/' + path
learn.load(name)
```

```python
learn.fine_tune(25)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```
